```python
# Let's check the dataset for any missing values
data.isnull().sum()
```

```python
# Create an imputer object
from sklearn.impute import SimpleImputer
# Handle missing values using mean imputation
imputer = SimpleImputer(strategy='mean')
data['SL'] = imputer.fit_transform(data[['SL']])
data['SW'] = imputer.fit_transform(data[['SW']])
data['PL'] = imputer.fit_transform(data[['PL']])
```

Encode the target variable:

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Classification'] = le.fit_transform(data['Classification'])
```

# Now, let's separate the features (SL, SW, PL, PW) and the target variable (Classification)

```python
X = data.drop('Classification', axis=1)
y = data['Classification']
```

Perform feature scaling:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# 3. Find out which classification model gives the best result
# We'll try different classification algorithms and evaluate their performance.


```python
# Importing the required libraries for classification models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Initializing the models
models = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
   
}
```

```python
# Training and evaluating each model
results = {}
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[model_name] = accuracy
```

```python
# Printing the results
for model_name, accuracy in results.items():
    print(f"{model_name}: {accuracy}")

```

```python
# The model with the highest accuracy will be considered the best model
best_model = max(results, key=results.get)
print(f"\nBest Model: {best_model}")
```
